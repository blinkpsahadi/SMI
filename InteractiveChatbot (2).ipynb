```python
!pip install openai sqlalchemy psycopg2-binary > /dev/null 2>&1
!pip install openai --upgrade > /dev/null 2>&1
!pip install biopython > /dev/null 2>&1
!pip install fpdf > /dev/null 2>&1

from openai import OpenAI
from fpdf import FPDF

# Initialiser le client OpenAI
client = OpenAI(
    base_url="https://integrate.api.nvidia.com/v1",
    api_key="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"
)

# Fonction pour exporter l'historique du chat en PDF
def export_chat_to_pdf(chat_history):
    pdf = FPDF()
    pdf.set_auto_page_break(auto=True, margin=15)
    pdf.add_page()
    pdf.set_font("Arial", size=12)

    pdf.cell(200, 10, txt="Historique du Chat", ln=True, align='C')
    pdf.ln(10)

    for message in chat_history:
        role = "Utilisateur" if message["role"] == "user" else "Chatbot"
        pdf.set_font("Arial", style='B', size=12)
        pdf.multi_cell(0, 10, txt=f"{role}:")
        pdf.set_font("Arial", size=12)
        pdf.multi_cell(0, 10, txt=message["content"])
        pdf.ln(5)

    pdf.output("Chat_Historique.pdf")
    print("\nL'historique du chat a été exporté en PDF : Chat_Historique.pdf")

# Fonction principale du chat
def start_chat():
    chat_history = []  # Pour stocker l'historique des messages

    while True:
        # Demander à l'utilisateur de saisir une question ou une demande
        user_input = input("Veuillez entrer votre demande : ")

        # Ajouter la demande de l'utilisateur à l'historique
        chat_history.append({"role": "user", "content": user_input})

        # Créer la demande dynamique avec l'input de l'utilisateur
        completion = client.chat.completions.create(
            model="meta/llama3-70b-instruct",
            messages=[{"role": "user", "content": user_input}],
            temperature=0.5,
            top_p=1,
            max_tokens=1024,
            stream=True
        )

        # Stocker la réponse du chatbot
        chatbot_response = ""
        print("Chatbot:", end=" ")
        for chunk in completion:
            if chunk.choices[0].delta.content is not None:
                print(chunk.choices[0].delta.content, end="")
                chatbot_response += chunk.choices[0].delta.content
        print()

        # Ajouter la réponse du chatbot à l'historique
        chat_history.append({"role": "chatbot", "content": chatbot_response})

        # Offrir la possibilité de continuer ou de quitter
        continue_chat = input("\nSouhaitez-vous poser une autre question ? (oui / non) : ").strip().lower()
        if continue_chat != 'oui':
            print("Merci d'avoir utilisé le chatbot ! À bientôt.")
            # Exporter l'historique du chat en PDF
            export_chat_to_pdf(chat_history)
            break

# Lancer le chat
if __name__ == "__main__":
    start_chat()
```
